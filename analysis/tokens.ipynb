```python
import pandas as pd
```

```python
df = pd.read_csv("../token_counts.csv")
df.head()

dfem = df[df["condition"] == "emotional"]
dfne = df[df["condition"] == "neutral"]
```

```python
        dfem.head()
        mean_em = dfem["n_tokens"].mean()
        std_em = dfem["n_tokens"].std()

        mean_ne = dfne["n_tokens"].mean()
        std_ne = dfne["n_tokens"].std()

        print(f"Emotional: mean={mean_em:.2f}, std={std_em:.2f}")
        print(f"Neutral: mean={mean_ne:.2f}, std={std_ne:.2f}")

        #t test to compare the means
        from scipy.stats import ttest_ind
        t_stat, p_value = ttest_ind(dfem["n_tokens"], dfne["n_tokens"], equal_var=False)
        print(f"T-statistic: {t_stat:.2f}, P-value: {p_value:.2f}")
```
